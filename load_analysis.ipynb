# **Exploratory Data Analysis of Hourly Electricity Load**


This notebook performs basic exploratory analysis on hourly electricity load data to understand demand patterns and variability.

**Data Loading**

The following cell loads the hourly electricity load dataset and performs basic inspection

```python
import pandas as pd
df = pd.read_csv("data/PJME_hourly.csv")
#converting column 1 to datetime format
df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.set_index('Datetime')
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
df.plot(legend=False)
plt.title("Hourly Electricity Load")
plt.xlabel("Time")
plt.ylabel("Load (MW)")
plt.tight_layout()
plt.show()
```

The plot shows long-term demand trends and significant temporal variability, with visible seasonal patterns.

```python

daily_load = df.resample('D').mean()

plt.figure(figsize=(10,4))
daily_load.plot(legend=False)
plt.title("Daily Average Load")
plt.xlabel("Time")
plt.ylabel("Load (MW)")
plt.tight_layout()
plt.show()
```

Daily aggregation smooths short-term fluctuations and highlights underlying seasonal demand behavior.
